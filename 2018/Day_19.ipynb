Therefore this is a program which finds the sum of the factors of either a large or an extremely large number.

```python
p1_target = 861
p2_target = 10_551_261
```

```python
def sum_of_factors(target):
    return target + sum(divisor for divisor in range(1, target//2) if target % divisor == 0)
```

```python
p1 = sum_of_factors(p1_target)
p2 = sum_of_factors(p2_target)
print(f'Part 1: {p1}')
print(f'Part 2: {p2}')
```
